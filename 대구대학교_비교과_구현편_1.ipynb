```python
def data_descriptions(data):

  cmax = np.max(data, axis=0)
  cmin = np.min(data, axis=0)
  cmean = np.mean(data, axis=0)
  cstd = np.std(data, axis=0)
  outlier_idx = np.where(data[:4]>100)[0]
  outlier1 = len(outlier_idx)

  print(f'인스턴스 계수는 {data.shape[0]}이고 변수의 개수는 {data.shape[1]}입니다.')
  print(f'x5 변수의 대한 이상치는 {outlier1}개 입니다.')

  return cmax, cmin, cmean, cstd
```

```python
data = np.random.normal(0,10, (100,1))
target = np.random.normal(0,1, 100)
```

```python
class LinearReg:

  def __init__(self, data, target, scaling=False):
    self.data = data
    self.target = target
    self.num_instances = self.data.shape[0]
    self.num_features = self.data.shape[1]
    self.scaling = scaling
    print(f'num_instances = {self.num_instances}, num_features = {self.num_features}, scaling = {scaling}')

  def minmax(self):
    for i in range(self.num_features):
      col = self.data[:,i]
      self.data[:,i] = (self.data[:,i]-np.min(col))/(np.max(col)-np.min(col))  

    return self

  def solver(self):
    if self.scaling == True:
      self.minmax()
    
    plt.plot(self.data, self.target, '.')
    plt.show()
```

```python
m1 = LinearReg(data, target)
m1.solver()
```

```python
#'''
#from torch.utils.data import Dataset

#class Mydataset(Dataset):

#  def __init__(self):  # 필요한 기본값

#  def __getitem__(self, index): # 뽑아낼 데이터

#  def __len__(self): # 클래스 내에 들어온 데이터 개수

#양식 저장!

#'''

from torch.utils.data import Dataset

class TensorData(Dataset):

  def __init__(self, x_data, y_data):
    self.x_data = torch.FloatTensor(x_data) # 이미지 데이터를 FloatTensor로 변형
    self.x_data = self.x_data.permute(0,3,1,2) # 이미지 너비 높이 채널 -> 배치크키 채널 너비 높이 (순서변경, 숫자는 위치)
    self.y_data = torch.LongTensor(y_data) # 라벨 데이터를 LongTensor로 변형
    self.len = self.y_data.shape[0]

  def __getitem__(self, index):
    return self.x_data[index], self.y_data[index] # 뽑아낼 데이터를 적어둔다

  def __len__(self):
    return self.len # 클래스 내의 들어온 데이터의 개수
```

```python
import torch
import torchvision.transforms as tr # 이미지 전처리 기능을 제공
from torch.utils.data import DataLoader, Dataset # 데이터를 모델에 사용할 수 있게 정리 해줌
import numpy as np
```

```python
# 데이터 불러오기 (glub -> PIL, openCV 등으로 불러오기 가능)

train_images = np.random.randint(256, size=(100,32,32,3)) # 이미지 높이 너비 채널
train_labels = np.random.randint(2, size=(100,1))
```

```python
# 전처리
from torch.utils.data import Dataset

class Mydataset(Dataset):

  def __init__(self, x_data, y_data, transform=None):  # 필요한 기본값

    self.x_data = x_data
    self.y_data = y_data
    self.transform = transform
    self.len = len(y_data)

  def __getitem__(self, index): # 뽑아낼 데이터
    sample = self.x_data[index], self.y_data[index]

    if self.transform:
      sample = self.transform(sample)

    return sample

  def __len__(self): # 클래스 내에 들어온 데이터 개수
    return self.len
```

```python
class ToTensor:
  def __call__(self, sample):
    inputs, labels = sample
    inputs = torch.FloatTensor(inputs) # 텐서로 변환
    inputs = inputs.permute(2,0,1) # 크기 변환 __call__에서는 한장씩 받기 때문에 4개가 아니라 3개
    return inputs, torch.LongTensor(labels) # 텐서로 변환
  
class LinearTensor:

  def __init__(self, slope=1, bias=0):
    self.slope = slope
    self.bias = bias

  def __call__(self, sample):
    inputs, labels = sample
    inputs = self.slope*inputs + self.bias # ax+b 계산하기
    return inputs, labels
```

```python
trans = tr.Compose([ToTensor(), LinearTensor(2,5)]) # 텐서 변환 후 선형식 2x+5 연산
dataset1 = Mydataset(train_images, train_labels, transform=trans)
train_loader1 = DataLoader(dataset1, batch_size=10, shuffle=True)
```

```python
dataiter = iter(train_loader1)
images1, labels1 = dataiter.next()
print(images1.size())
```

커스텀 데이터 + torchvision.transforms 전처리

```python
# torchvision.transforms 는 입력 이미지가 일반적으로 tensor, PILImage 타입이여야 한다.

class MyTransform:

  def __call__(self, sample):
    inputs, labels = sample
    inputs = torch.FloatTensor(inputs)
    inputs = torch.permute(2,0,1)
    labels = torch.FloatTensor(labels)

    transf = tr.Compose([tr.ToPILImage(), tr.Resize(128), tr.ToTensor(), tr.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))])
    final_output = transf(inputs)
  
    return final_output, labels
```

```python
dataset2 = Mydataset(train_images, train_labels, transform=MyTransform())
train_loader2 = DataLoader(dataset2, batch_size=10, shuffle=True)
```

```python
# 텐서 데이터로 변환하는 클래스

class TensorData(Dataset):

  def __init__(self, x_data, y_data):
    self.x_data = torch.FloatTensor(x_data) # 이미지 데이터를 FloatTensor로 변형
    self.y_data = torch.FloatTensor(y_data) # 라벨 데이터를 LongTensor로 변형
    self.len = self.y_data.shape[0]

  def __getitem__(self, index):
    return self.x_data[index], self.y_data[index] # 뽑아낼 데이터를 적어둔다

  def __len__(self):
    return self.len # 클래스 내의 들어온 데이터의 개수
```

모델 구축 

모델은 Regressor로 정의하며 입력층(노드 13개), 2개의 은닉층(50, 30개), 출력층(1개) 구성한다. 데이터의 변수가 13개 이므로 입력층 노드 13개, 출력층은 집 값이므로 1개, 은닉층에 대해서는 실험을 하면서 튜닝할 수 있다.

```python
class Regressor(nn.Module):
  def __init__(self):
    super().__init__() # 모델 연산 정의
    self.fc1 = nn.Linear(13, 50, bias=True) # 입력층(13) -> 은닉층1(50)으로 가는 연산
    self.fc2 = nn.Linear(50, 30, bias=True) # 은닉층1(50) -> 은닉층2(30)으로 가는 연산
    self.fc3 = nn.Linear(30, 1, bias=True) # 은닉층2(30) -> 출력층(1)으로 가는 연산
    self.dropout = nn.Dropout(0.2) #연산이 될 때마다 20%의 비율로 랜덤하게 노드를 없앤다.

  def forward(self, x): # 모델 연산의 순서를 정의
    x = F.relu(self.fc1(x)) # Linear 계산 후 활성화 함수 ReLU를 적용
    x = self.dropout(F.relu(self.fc2(x))) # 은닉층2에서 드랍아웃을 적용 (30개의 20% 6개의 노드가 제외)
    x = F.relu(self.fc3(x)) # Linear 계산 후 활성화 함수 ReLU를 적용

    return x

# 드랍아웃은 과적합(overfitting)을 방지하기 위해 노드의 일부를 배체하고 계산하는 방식, 절대로 출력층에 사용해서는 안됨
```

모델, 손실함수, 최적화 방법 선언

```python
model = Regressor()
criterion = nn.MSELoss()

# lr은 학습률
# weight_decey는 L2 정규화에서의 penalty 정도를 의미
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-7)
```

```python
loss_ = [] # 그래프를 그리기 위한 loss 저장용 리스크
n = len(trainloader) # 배치 개수
 
for epoch in range(400): # 400번 학습
  running_loss = 0.0

  for i, data in enumerate(trainloader, 0): # 무작위로 섞인 32개 데이터가 있는 배치가 하나 씩 들어있다. # enumerate는 i에 trainloader 데이터 index가 뽑아 나온다.

    inputs, values = data # 데이터에는 X, Y가 있다.

    optimizer.zero_grad() # 최적화 초기화

    outputs = model(inputs) # 모델에 입력값 대입 후 예측값 산출
    loss = criterion(outputs, values) # 손실함수 계산
    loss.backward() # 손실함수 기준으로 역전파 설정
    optimizer.step() # 역전파를 진행하고 가중치 업데이트

    running_loss += loss.item()

  loss_.append(np.sqrt(running_loss) / n) # RMSE 계산

print('Finished Traning')
```

모델 평가


```python
def evaluation(dataloader):

  predictions = torch.tensor([], dtype=torch.float) # 예측값을 저장하는 텐서
  actual = torch.tensor([], dtype=torch.float) # 실제값을 저장하는 텐서

  with torch.no_grad(): # grad()를 정의안해도 된다.
    model.eval() # 평가할 때에는 .eval()을 반드시 사용해야 한다. (dropout이 작동을 안함)
    for data in dataloader:
      inputs, values = data
      outputs = model(inputs)

      predictions = torch.cat((predictions, outputs), 0) # cat을 통해 예측값을 누적 (tensor이용이 필요없을때에는 list, numpy arrary 또는 일반 숫자값으로 쌓아도 된다.)
      actual = torch.cat((actual, values), 0) # cat을 통해 실제값을 누적

  predictions =  predictions.numpy() # 넘파이 배열로 변형
  actual = actual.numpy() # 넘파이 배열로 변형
  mse = np.sqrt(mean_squared_error(predictions, actual)) # sklearn으로 RMSE 계산 

  return mse
```

```python
train_mse = evaluation(trainloader)
test_mse = evaluation(testloader)

print('Train_MSE :', train_mse)
print('Test_MSE :', test_mse)
```

CNN 모델 구축하기

```python
class Net(nn.Module):
  def __init__(self):
    super(Net, self).__init__()
    self.conv1 = nn.Conv2d(3, 6, 5) # 합성곱 연산 (입력 채널수 3, 출력 채널수 6, 필터크기 5x5, stride=1(defualt))
    self.pool1 = nn.MaxPool2d(2, 2) # 합성곱 연산 (필터크기 2, stride=2)
    self.conv2 = nn.Conv2d(6, 16, 5) # 합성곱 연산 (입력 채널수 6, 출력 채널수 16, 필터크기 5x5, stride=1(defualt))
    self.pool2 = nn.MaxPool2d(2, 2) # 합성곱 연산 (필터크기 2, stride=2)
    self.fc1 = nn.Linear(16 * 5 * 5, 120) # 5x5 피쳐맵 16개를 일렬로 피면 16*5*5개의 노드가 생성, 120개의 노드를 가진 히든레이어 생성
    self.fc2 = nn.Linear(120, 10) # 120개의 노드에서 클래스의 개수인 10개의 노드로 연산

  def forward(self, x):
    x = self.pool1(F.relu(self.conv1(x))) # conv1 -> ReLU -> pool1
    x = self.pool2(F.relu(self.conv2(x))) # conv2 -> ReLU -> pool2
    x = x.view(-1, 16 * 5 * 5) # 5x5 피쳐맵 16개를 일렬로 만든다.
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))

    return x

net = Net().to(device) # 모델 선언
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=1e-3, momentum=0.9)
```

```python
loss_ = [] # 그래프를 그리기 위한 loss 저장용 리스트
n = len(trainloader) # 배치 수

for epoch in range(10): # 10번 학습 한다.
  
  running_loss = 0.0
  for i, data in enumerate(trainloader, 0):
    
    inputs, labels = data[0].to(device), data[1].to(device) # 배치 데이터 .to(device) : GPU용 데이터로 변환
    optimizer.zero_grad()

    outputs = net(inputs) # 예측 값 산출
    loss = criterion(outputs, labels) # 손실함수 계산
    loss.backward() # 손실함수 기준으로 역전파 선언
    optimizer.step() # 가중치 최적화

    # print statistics
    running_loss += loss.item()

  loss_.append(running_loss / n)
  print('[%d], loss : %.3f' %(epoch + 1, running_loss / len(trainloader)))

print('Finished Training')
```

```python
class VanillaRNN(nn.Module):
  def __init__(self, input_size, hidden_size, sequence_length, num_layers, device):
    super(VanillaRNN, self).__init__()
    self.device = device
    self.hidden_size = hidden_size
    self.num_layers = num_layers
    self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True) # 원래 pytorch.RNN은 sequence_length가 먼저 나와야 하지만 위치를 바꾸기 귀찮을때 
                                                                             # batch_first=True를 통해 안바꿔도 된다. (현 데이터는 batch_size, sequence_lenght, value_size 순)

    self.fc = nn.Sequential(nn.Linear(hidden_size*sequence_length, 1), nn.Sigmoid()) # many to many

  def forward(self, x):
    h0 = torch.zeros(self.num_layers, x.size()[0], self.hidden_size).to(self.device) # 초기 hidden state 설정
    out, _ = self.rnn(x, h0) # out : RNN의 마지막 레이어로 부터 나온 output feature 반환, hn : hidden state로 변환
    out = out.reshape(out.shape[0], -1) # many to many 전략
    out = self.fc(out)
    return out
```

```python
model = VanillaRNN(input_size=input_size,
                   hidden_size=hidden_size,
                   sequence_length=sequence_length,
                   num_layers=num_layers,
                   device=device).to(device)
```

```python
criterion = nn.MSELoss()

lr = 1e-3
num_epoch = 200
optimizer = optim.Adam(model.parameters(), lr=lr)
```

```python
class LSTM(nn.Module):

  def __init__(self, input_size, hidden_size, sequence_length, num_layers, device):
    super(LSTM, self).__init__()
    self.device = device
    self.hidden_size = hidden_size
    self.num_layers = num_layers
    self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
    self.fc = nn.Linear(hidden_size*sequence_length, 1)

  def forward(self, x):
    h0 = torch.zeros(self.num_layers, x.size()[0], self.hidden_size).to(self.device)
    c0 = torch.zeros(self.num_layers, x.size()[0], self.hidden_size).to(self.device) # cell state 추가
    out, _ = self.lstm(x, (h0, c0)) # output, (hn, cn) : cell state와 hidden state만 반환, 순서쌍으로 만듬 (h0, c0)
    out = out.reshape(out.shape[0], -1) # state 추가
    out = self.fc(out)
    return out  
```

GRU (LSTM의 간소한 버젼)

기본 RNN과 비슷하게 하면 된다. (nn.RNN -> nn.GRU)

GRU가 가장 효과적이다.

```python
class GRU(nn.Module):
  def __init__(self, input_size, hidden_size, sequence_length, num_layers, device):
    super(GRU, self).__init__()
    self.device = device
    self.hidden_size = hidden_size
    self.num_layers = num_layers
    self.rnn = nn.GRU(input_size, hidden_size, num_layers, batch_first=True) # 원래 pytorch.RNN은 sequence_length가 먼저 나와야 하지만 위치를 바꾸기 귀찮을때 
                                                                             # batch_first=True를 통해 안바꿔도 된다. (현 데이터는 batch_size, sequence_lenght, value_size 순)

    self.fc = nn.Sequential(nn.Linear(hidden_size*sequence_length, 1), nn.Sigmoid()) # many to many

  def forward(self, x):
    h0 = torch.zeros(self.num_layers, x.size()[0], self.hidden_size).to(self.device) # 초기 hidden state 설정
    out, _ = self.rnn(x, h0) # out : RNN의 마지막 레이어로 부터 나온 output feature 반환, hn : hidden state로 변환
    out = out.reshape(out.shape[0], -1) # many to many 전략
    out = self.fc(out)
    return out
```

오코인코더 모델 생성


```python
class Flatten(nn.Module): # 4D -> 2D로 계산하기
  def forward(self, x):
    batch_size = x.shape[0]
    return x.view(batch_size, -1) # (배치 수, 채널 수, 이미지 너비, 이미지 높이) -> (배치 수, 채널 수*이미지 너비*이미지 높이)

class Deflatten(nn.Module): # 2D -> 4D로 계산하기

  def __init__(self, k):
    super(Deflatten, self).__init__()
    self.k = k
  
  def forward(self, x):
    s = x.size()

    # 벡터 사이즈 = 채널 수*이미지 너비*이미지 높이
    # 벡터 사이즈 = 채널 수*이미지 사이즈**2
    # 이미지 사이즈 = (벡터 사이즈//채널 수)**.5
    feature_size = int((s[1]//self.k)**.5)

    return x.view(s[0], self.k, feature_size, feature_size) # (배치 수, 채널 수*이미지 너비*이미지 높이) -> (배치 수, 채널 수, 이미지 너비, 이미지 높이)

class Autoencoder(nn.Module):
  def __init__(self):
    super(Autoencoder, self).__init__()

    k = 16
    self.encoder = nn.Sequential(
        nn.Conv2d(1, k, 3, stride=2),
        nn.ReLU(),
        nn.Conv2d(k, 2*k, 3, stride=2),
        nn.ReLU(),
        nn.Conv2d(2*k, 4*k, 3, stride=1),
        nn.ReLU(),
        Flatten(), # 일렬로 핀다.
        nn.Linear(1024, 10),
        nn.ReLU()
    )
    # ConvTranspose2d
    # 입력 성분(Cov의 결과)을 출력 성분(Cov의 입력)으로 미분하여 그 값을 입력 벡터와 곱해 출력 벡터를 산출한다. 데이터 크기를 늘린다.
    # 출력 전 벡터는 행렬 형태로 변환한다.
    self.decoder = nn.Sequential(
        nn.Linear(10, 1024),
        nn.ReLU(),
        Deflatten(4*k), # 다시 정사각형 행렬로 변환
        nn.ConvTranspose2d(4*k, 2*k, 3, stride=1), # (입력 채널 수, 출력 채널 수, 필터 크기, stride)
        nn.ReLU(),
        nn.ConvTranspose2d(2*k, k, 3, stride=2),
        nn.ReLU(),
        nn.ConvTranspose2d(k, 1, 3, stride=2, output_padding=1),
        nn.Sigmoid()
    )

  def forward(self, x):
    encoder = self.encoder(x)
    decoder = self.decoder(encoder)

    return decoder
```

```python
class Generator(nn.Module): # 이미지 생성 클래스
  def __init__(self):
    super(Generator, self).__init__()
    self.n_feature = 128 # 지정해주면 된다. 
    self.n_out = 784 # FasionMNIST의 이미지가 28*28 이다. (고정 값)
    self.linear = nn.Sequential(
        nn.Linear(self.n_feature, 256), # 지금 모델은 Linear GAN, Linear 대신 ConvTranspose를 사용하면 Conv GAN(DCGAN)이다.
        nn.LeakyReLU(0.2), # 0.2는 음수값에 양의 기울기 (왼쪽 식 : y = 0.2x, 오른쪽 식 : y = x)
        nn.Linear(256, 512),
        nn.LeakyReLU(0.2),
        nn.Linear(512, 1024),
        nn.LeakyReLU(0.2),
        nn.Linear(1024, self.n_out),
        nn.Tanh()
    )

  def forward(self, x):
    x = self.linear(x)
    x = x.view(-1, 1, 28, 28) # 28 * 28로 변환
    return x

class Discriminator(nn.Module): # 진짜인지 가짜인지 판별하는 클래스
  def __init__(self):
    super(Discriminator, self).__init__()
    self.n_in = 784 
    self.n_out = 1
    self.linear = nn.Sequential(
        nn.Linear(self.n_in, 1024), # DC GAN이라면 Linear 대신에 Conv2d를 사용한다.
        nn.LeakyReLU(0.2), # 다잉뉴런 현상을 방지하기 위해 LeakyReLU 사용 (다잉뉴런 : ReLU는 음수가 0, 어떠한 입력값이 와도 활성화 값이 0이되는 현상, 업데이트가 안됨)
        nn.Dropout(0.3),
        nn.Linear(1024, 512),
        nn.LeakyReLU(0.2),
        nn.Dropout(0.3),
        nn.Linear(512, 256),
        nn.LeakyReLU(0.2),
        nn.Dropout(0.3),
        nn.Linear(256, self.n_out),
        nn.Sigmoid() # Discriminator은 진짠지 가짠지를 구별하는 것이므로 Sigmoid를 사용해 0~1 값으로 만든다.
    )
  
  def forward(self, x):
    x = x.view(-1, self.n_in) # 이미지를 펴줘야 한다. 28 * 28 = 784
    x = self.linear(x)
    return x
```

손실함수 및 최적화 방법 정의

```python
generator = Generator().to(device)
discriminator = Discriminator().to(device)

pretrained = False
if pretrained == True:
  discriminator.load_state_dict(torch.load('./models/fmnist_disc.pth'))
  generator.load_state_dict(torch.load('./models/fmnist_gner/pth'))

g_optim = optim.Adam(generator.parameters(), lr=2e-4) # generator, discriminator 둘다 업데이트를 해야해서 optim이 두 버젼으로 있다.
d_optim = optim.Adam(discriminator.parameters(), lr=2e-4)

g_losses = []
d_losses = []
images = [] # GIF 형태로 만들어 동영상으로 만들어 보자

criterion = nn.BCELoss() # Boundary Cross Entropy Loss

def noise(n, n_features = 128): # input 값
  return Variable(torch.randn(n, n_features)).to(device)

def label_ones(size): # 라벨 : 진짜면 1
  data = Variable(torch.ones(size, 1))
  return data.to(device)

def label_zeros(size): # 라벨 : 가짜면 0
  data = Variable(torch.zeros(size, 1))
  return data.to(device)
```

학습 전략 정의

```python
def train_discriminator(optimizer, real_data, fake_data): # discriminator 학습하기
  n = real_data.size(0)

  optimizer.zero_grad()

  prediction_real = discriminator(real_data)
  d_loss = criterion(prediction_real, label_ones(n)) # 진짜 이미지에 대한 loss
  d_loss.backward()

  prediction_fake = discriminator(fake_data)
  g_loss = criterion(prediction_fake, label_zeros(n)) # 가짜 이미지에 대한 loss
  g_loss.backward()

  optimizer.step()

  return d_loss + g_loss # loss 더하기

def train_generator(optimizer, fake_data):
  n = fake_data.size(0)
  optimizer.zero_grad()

  prediction = discriminator(fake_data)
  loss = criterion(prediction, label_ones(n)) # 가짜 이미지를 진짜로 하고 loss를 계산해 loss가 작아지면 discriminator가 구별을 잘 못한는 뜻이다.

  loss.backward()
  optimizer.step()

  return loss
```

학습하기

```python
num_epochs = 201
test_noise = noise(64) # 128개의 벡터를 64개 생성

l = len(trainloader)

for epoch in range(num_epochs):
  g_loss = 0.0
  d_loss = 0.0

  for data in trainloader:
    imgs, _ = data # 비지도 학습이라 label을 불러오지 않는다.
    n = len(imgs)

    fake_data = generator(noise(n)).detach() # 가짜이미지 생성
    real_data = imgs.to(device)
    d_loss += train_discriminator(d_optim, real_data, fake_data) # 진짜 vs 가짜 
    fake_data = generator(noise(n)) # 가짜이미지 생성
    g_loss += train_generator(g_optim, fake_data) # 가짜를 더 진짜럼 만듬

  img = generator(test_noise).cpu().detach()
  img = make_grid(img)
  images.append(img)
  g_losses.append(g_loss/l)
  d_losses.append(d_loss/l)

  if epoch % 10 == 0:
    print('Epoch {}: g_loss: {:.3f} d_loss: {:.3f}/r'.format(epoch, g_loss/l, d_loss/l))
    torch.save(discriminator.state_dict(), './model/fmnist_disc.pth')
    torch.save(generator.state_dict(), './model/fmnist_gner.pth')

print('Training Finished')
```
